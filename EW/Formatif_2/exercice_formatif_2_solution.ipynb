```python
#@title Importations
import numpy as np
import matplotlib.pyplot as plt
```

```python
#@title Fonctions

# Retourne le rayon en fonction de l'aire
def calculer_rayon(aire):
  return np.sqrt(aire / np.pi)

# Fait le calcul étrange demandé dans l'énoncé. Le résultat doit
# être converti en int.
def calculer_bizarro(x, y):
  resultat = x+y + x-y + x*y + x/y
  # ou ...
  # resultat = 2x + x*y + x/y
  return int(resultat)

# Applique le rabais (en %) et retourne le montant après rabais
def appliquer_rabais(montant, pourcentage_rabais):
  rabais = montant * pourcentage_rabais / 100
  montant_apres_rabais = montant - rabais
  return montant_apres_rabais

# Ajoute la tps de 5% et la tvq de 9.975% au montant et retourne le montant
# total incluant les taxes.
def ajouter_taxes(montant):
  tps = montant * 0.05
  tvq = montant * 0.09975
  montant_apres_taxe = montant + tps + tvq
  return montant_apres_taxe

# Retourne la valeur du polynôme ax²+bx + c en position x.
def polynome_degre_2(a, b, c, x):
  return a*x**2 + b*x + c

# Retourne la pente du polynôme ax²+bx+c entre les points x et x + delta_x
def calculer_pente(a, b, c, x, delta_x):
  valeur_0 = polynome_degre_2(a, b, c, x)
  valeur_1 = polynome_degre_2(a, b, c, x + delta_x)
  pente = (valeur_1 - valeur_0) / delta_x
  return pente

# Retourne la position (sur l'axe des x) en appliquant la formule spécifiée
# dans l'énoncé
def calculer_position(v_0, angle_tir, x_0, t):
    position = v_0*np.cos(np.radians(angle_tir))*t + x_0
    return position

# Retourne la hauteur (sur l'axe des y) en appliquant la formule spécifiée
# dans l'énoncé
def calculer_hauteur(v_0, angle_tir, y_0, t):
    angle_rad = np.radians(angle_tir)
    g = 9.81
    hauteur = -0.5*g*t**2 + v_0*np.sin(angle_rad)*t + y_0
    return hauteur
```

```python
#@title Programme 1

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p1(rayon):
  print(f"Le rayon du cercle est {rayon:.4f}")

# Écrire votre code apres cette ligne

aire = float(input("Saisir l'aire: "))
rayon = calculer_rayon(aire)

afficher_resultat_p1(rayon)
```

```python
#@title Programme 2

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p2(x, y, resultat_calcul):
  print(f"{x} bizarro {y} = {resultat_calcul}")

# Écrire votre code après cette ligne
x = float(input("Écrire la première valeur (x): "))
y = float(input("Écrire la deuxième valeur (y): "))

afficher_resultat_p2(x, y, calculer_bizarro(x, y))
```

```python
#@title Programme 3

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p3(montant_final):
  print(f"Le montant final est {montant_final:.2f}")

# Écrire votre code après cette ligne
montant_base = float(input("Saisir le montant: "))
rabais = int(input("Saisir le pourcentage du rabais (sans le %): "))
montant_avec_rabais = appliquer_rabais(montant_base, rabais)
montant_final = ajouter_taxes(montant_avec_rabais)
afficher_resultat_p3(montant_final)
```

```python
#@title Programme 4

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p4(jours, heures, minutes):
  print(f"{jours} jour(s), {heures} heure(s) et {minutes} minute(s)")

# Écrire votre code après cette ligne
nombre_degres = float(input("Entrez le nombre de degrés de déplacement : "))
minutes_totales = int(nombre_degres / 0.5)
jours = minutes_totales // (24*60)
heures = (minutes_totales % (24*60)) // 60
minutes = minutes_totales % 60

afficher_resultat_p4(jours, heures, minutes)
```

```python
#@title Programme 5

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p5(pente):
  print(f"La pente est {pente:.5g}")

# Écrire votre code apres cette ligne
a = float(input("Entrez a: "))
b = float(input("Entrez b: "))
c = float(input("Entrez c: "))
x = float(input("Entrez x: "))
delta_x = float(input("Entrez delta_x: "))

pente = calculer_pente(a, b, c, x, delta_x)
afficher_resultat_p5(pente)
```

```python
#@title Programme 6

# Cette fonction ne doit pas être modifiée mais utilisée pour afficher les résultats
def afficher_resultat_p6(t, distance, hauteur):
  print(f"t = {t} s => p = {distance:.2f} m, h = {hauteur:.2f} m")

# Écrire votre code après cette ligne


y_0 = 1.50
v_0 = 9.0
x_0 = 0.0
t = float(input("Entrez la valeur de t: "))
angle_tir = 26

distance = calculer_position(v_0, angle_tir, x_0, t)
hauteur = calculer_hauteur(v_0, angle_tir, y_0, t)

afficher_resultat_p6(t, distance, hauteur)
```
